# Data analysis (EDA)

#Results shows that females ratio is greater than males 

```python
ax = sns.countplot(x="Gender",data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

#Results shows that females ratio is greater than males 

```python
ax=sns.countplot(x="Age Group",data=df,hue = "Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
age_group = df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data = age_group)
```

#26-35 males and females are more athenthic customers

# state

```python
df.columns
```

```python
state_order=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
sns.barplot(data=state_order,x="State",y="Orders")
```

```python
state_amount=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
sns.barplot(data=state_amount,x="State",y="Amount")
```

# Marital_status

```python
ax=sns.countplot(data=df,x="Marital_Status")
sns.set(rc={"figure.figsize":(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
marital_status=df.groupby(["Marital_Status"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(10,5)})
sns.barplot(data=marital_status,x="Marital_Status",y="Amount")
```

```python
#most are married females
```

# Occupation

```python
ax=sns.countplot(data=df,x="Occupation")
sns.set(rc={"figure.figsize":(15,7)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occupation=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(10,5)})
sns.barplot(data=occupation,x="Occupation",y="Amount")
```

```python
#most buyers belond to IT sector Healthcare and Aviation
```

# product catogery

```python
ax=sns.countplot(data=df,x="Product_Category")
sns.set(rc={"figure.figsize":(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
product_cat=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,10)})
sns.barplot(data=product_cat,x="Product_Category",y="Amount")
```

```python
#most product beloongs to Food clothing and Electronics
```

```python
product_id=df.groupby(["Product_ID"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,10)})
sns.barplot(data=product_id,x="Product_ID",y="Amount")
```
